Dado que 60% dos emails são spam, queremos calcular quantos dos 963 emails na caixa de spam são não-spam. Usamos o Teorema de Bayes para encontrar a probabilidade de um email ser **não-spam** dado que ele foi marcado como spam.

$$
P(NS|M) = \frac{P(M|NS) \cdot P(NS)}{P(M)}
$$

Onde:

- P(S) = 0.60 é a probabilidade de um email ser spam.
- P(NS) = 0.40 é a probabilidade de um email ser não-spam.
- P(M|S) = 0.98 é a probabilidade do filtro marcar corretamente um email spam como spam.
- P(M|NS) = 0.02 é a probabilidade do filtro marcar incorretamente um email não-spam como spam.

Primeiro, calculamos P(M), a probabilidade de um email ser marcado como spam:

$$
P(M) = P(M|S) \cdot P(S) + P(M|NS) \cdot P(NS)
$$

Substituindo os valores:

$$
P(M) = 0.98 \cdot 0.60 + 0.02 \cdot 0.40
$$

$$
P(M) = 0.588 + 0.008 = 0.596
$$

Agora, aplicamos o Teorema de Bayes para encontrar \(P(NS|M)\):

$$
P(NS|M) = \frac{0.02 \cdot 0.40}{0.596}
$$

$$
P(NS|M) = \frac{0.008}{0.596} \approx 0.01342
$$

Portanto, a probabilidade de um email ser **não-spam**, dado que ele foi marcado como spam, é aproximadamente 1.34%.

Agora, calculamos o número de emails não-spam entre os 963 emails marcados como spam:

$$
963 \cdot 0.01342 \approx 12.93
$$

Assim, cerca de **13 emails** dos 963 na caixa de spam provavelmente são **não-spam**.

- Precisamos de pelo menos 3 bits para codificar 6 valores diferentes $(2^3 = 8)$, que é a menor potência de 2 que pode acomodar 6 valores.
- Existem 36 caracteres no total em $v_1$.

Para $v_2$:

- $v_2$ contém 9 caracteres únicos: $a, d, f, g, h, j, k, n, s$.
- Precisamos de pelo menos 4 bits para codificar 9 valores diferentes $(2^4 = 16)$, que é a menor potência de 2 que pode acomodar 9 valores.
- Existem 36 caracteres no total em $v_2$.

**Portanto, para codificar $v_1$ e $v_2$ independentemente:**

- Para $v_1$: $3 \, \text{bits} \times 36 \, \text{caracteres} = 108 \, \text{bits}$.
- Para $v_2$: $4 \, \text{bits} \times 36 \, \text{caracteres} = 144 \, \text{bits}$.
- **Total**: $108 + 144 = 252 \, \text{bits}$.

### Codificando $v_1$ sabendo $v_2$:

Se soubéssemos $v_2$, poderíamos usar essa informação para codificar $v_1$ de forma mais eficiente. Vamos analisar a relação entre $v_1$ e $v_2$:

- Quando $v_2$ é 'a', $v_1$ é sempre 'a'.
- Quando $v_2$ é 's', $v_1$ é sempre 'e'.
- Quando $v_2$ é 'f', $v_1$ é sempre 'f'.
- Para outros caracteres em $v_2$, $v_1$ pode variar.

Podemos usar essas informações para criar uma codificação mais eficiente para $v_1$:

- Se $v_2$ for 'a', 's' ou 'f', não precisamos codificar $v_1$ (0 bits).
- Para outros casos, precisamos codificar qual dos 6 caracteres possíveis $v_1$ poderia ser (3 bits).

- 's' aparece 5 vezes.
- 'f' aparece 3 vezes.
- Outros caracteres aparecem 24 vezes.

$$
(0 \, \text{bits} \times 4) + (0 \, \text{bits} \times 5) + (0 \, \text{bits} \times 3) + (3 \, \text{bits} \times 24) = 72 \, \text{bits}
$$

### Economia na codificação de $v_1$:

- Codificação original: $108 \, \text{bits}$.
- Nova codificação: $72 \, \text{bits}$.
- **Economia**: $108 - 72 = 36 \, \text{bits}$.

Portanto, se soubéssemos $v_2$, economizaríamos 36 bits na codificação de $v_1$, o que representa uma redução de 33,33\% no número de bits necessários para codificar $v_1$.

````python
# se for um clique coupling, todos os nós devem ter mais do que um coupling edge
if min(node_cedges.values()) > 1:
   print("Clique coupling")
````

````python
# se for um star coupling, todos os nós devem ter um edge, ou |L| - 1 coupling edges
if set([1, layers - 1]) == set(node_cedges.values()):
   print("Star coupling")
````

````python
import networkx as nx
from itertools import combinations

# carrega os dados
hyperedges = []
with open("./cap4q3.txt", 'r') as f:
   for line in f:
      hyperedges.append(line.strip().split())

# transforma em um grafo unipartido
G_uni = nx.Graph()
for hyperedge in hyperedges:
   for combination in combinations(hyperedge, 2):
      G_uni.add_edge(*combination)

# transforma em um grafo bipartido
G_bip = nx.Graph()
hyperid = 1
for hyperedge in hyperedges:
   for node in hyperedge:
      G_bip.add_edge(node, "h%s" % hyperid)
   hyperid += 1
````
